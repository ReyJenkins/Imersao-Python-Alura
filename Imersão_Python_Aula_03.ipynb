# Aula 3 - Manipular dados com Pandas

```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel ("/content/Minha Solução Reinaldo - Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_de_acoes = pd.read_excel ("/content/Minha Solução Reinaldo - Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_total_de_acoes.head(10)
```

```python
df_ticker = pd.read_excel("/content/Minha Solução Reinaldo - Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/Minha Solução Reinaldo - Imersão Python - Tabela de ações.xlsx", sheet_name="Chatgpt")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_de_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (fundação ou início de operações)':'Idade'}).copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (em anos)':'Ano_de_fundacao'}).copy()
df_principal
```

```python
df_principal['Cat_Idade'] = df_principal['Ano_de_fundacao'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
# Calculando o maior valor
maior = df_principal['Variacao_rs'].max()

# Calculando o menor valor
menor = df_principal['Variacao_rs'].min()

# Calculando a média
media = df_principal['Variacao_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
#df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
#df_analise_saldo

# Agrupar e somar as variações por resultado
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()

# Ordenar o DataFrame pela coluna 'Variacao_rs'
df_analise_saldo = df_analise_saldo.sort_values(by='Variacao_rs')

# Exibir o DataFrame ordenado
df_analise_saldo
```

```python
#fig = px.bar(df_analise_saldo, x = 'Resultado', y = 'Variacao_rs', text = 'Variacao_rs', title = 'Variação Reais por Resultado')
#fig.show()

# Formatar os valores da coluna 'Variacao_rs' com o símbolo 'R$' e 2 casas decimais
df_analise_saldo['Variacao_rs'] = df_analise_saldo['Variacao_rs'].map('R${:.2f}'.format)

# Criar o gráfico com os valores formatados
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Resultado')
fig.update_traces(texttemplate='%{text}', textposition='outside')

fig.show()
```

```python
# Agrupar e somar as variações por segmento
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()

# Criar o gráfico de pizza com os valores agrupados por segmento
fig = px.pie(df_analise_segmento, values='Variacao_rs', names='Segmento', title='Variação Reais por Segmento')

# Exibir o gráfico
fig.show()
```

```python
# Calcular a coluna 'Análise Faixa Etária' usando a função unique do Pandas
analise_faixa_etaria = df_principal['Cat_Idade'].unique()

# Calcular a soma da 'Variação R$' para cada faixa etária usando a função sum e loc do Pandas
soma_variacao = []
for faixa_etaria in analise_faixa_etaria:
    soma = df_principal.loc[df_principal['Cat_Idade'] == faixa_etaria, 'Variacao_rs'].sum()
    soma_variacao.append(soma)

# Calcular a contagem da 'Quantidade de empresas' para cada faixa etária usando a função count e loc do Pandas
contagem_empresas = []
for faixa_etaria in analise_faixa_etaria:
    contagem = df_principal.loc[df_principal['Cat_Idade'] == faixa_etaria, 'Nome'].count()
    contagem_empresas.append(contagem)

# Criar um novo DataFrame com os resultados corrigidos
df_resultado = pd.DataFrame({
    'Análise Faixa Etária': analise_faixa_etaria,
    'Variação R$': soma_variacao,
    'Quantidade de empresas': contagem_empresas
})

# Exibir o DataFrame com os resultados corrigidos
df_resultado
```

```python
# Agrupar os dados pelo campo 'Cat_idade' e calcular a soma da 'Variacao_rs'
df_analise_cat_idade = df_principal.groupby('Cat_Idade')['Variacao_rs'].sum().reset_index()

# Ordenar o DataFrame pela coluna 'Variacao_rs' do menor para o maior
df_analise_cat_idade = df_analise_cat_idade.sort_values(by='Variacao_rs', ascending=True)

# Formatar os valores da coluna 'Variacao_rs' para duas casas decimais
df_analise_cat_idade['Variacao_rs'] = df_analise_cat_idade['Variacao_rs'].round(2)

# Criar o gráfico de barras com Plotly Express
fig = px.bar(df_analise_cat_idade, x='Cat_Idade', y='Variacao_rs', text='Variacao_rs',
             title='Variação Reais por Categoria de Idade',
             labels={'Cat_idade': 'Categoria de Idade', 'Variacao_rs': 'Variação R$'})

# Exibir o gráfico
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.show()
```
